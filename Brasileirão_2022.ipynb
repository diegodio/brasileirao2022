```python
#Other imports here
import os
import wget

import pandas as pd
from matplotlib import pyplot as plt
```

```python
driver = webdriver.Chrome('C:/Users/User/Desktop/chromedriver.exe')
driver.get("https://brasileirao.cbf.com.br/jogos-da-rodada/1")
```

```python
listaTimes = ["FLU", "SAN", "ACG", "FLA", "PAL", "CEA", "CFC", "GOI", "CAM", "INT", "BOT", "COR", "FOR", "CUI", "SAO", "CAP", "AVA", "AME", "JUV", "BRA"]
teams = {}
for i in listaTimes:
              teams[i] = {"id": i, "pontuacao": 0, "vitorias": 0, "empates":0, "derrotas": 0, "saldoDeGols":0, "golsPro": 0, "golsContra": 0, "golsProEmCasa": 0,"golsProFora": 0, "golsContraEmCasa": 0, "golsContraFora": 0}
        
for j in range (1,39):
    driver.get("https://brasileirao.cbf.com.br/jogos-da-rodada/"+str(j))
    getnomestimes=driver.find_elements_by_class_name("time_nome")
    lista_nome_times = []
    for i in getnomestimes:
        lista_nome_times.append(i.text)

    getgols=driver.find_elements_by_class_name("gols")
    lista_gols = []
    for i in getgols:
        lista_gols.append(i.text)
        
    if lista_gols[0] == "-\n-":
            break  
            
    print(f"{str(j)}ª RODADA")
    for i in range (0,10):
        print(lista_nome_times[i*2] + " " + lista_gols[i][0] + " x " + lista_gols[i][2] + " " + lista_nome_times[i*2+1])
        for j in lista_nome_times:
            if lista_nome_times[i*2] == j:
                teams[j]["golsProEmCasa"] += int(lista_gols[i][0])
                teams[j]["golsContraEmCasa"] += int(lista_gols[i][2])
                if int(lista_gols[i][0]) > int(lista_gols[i][2]):
                    teams[lista_nome_times[i*2]]["vitorias"] += 1
                    teams[lista_nome_times[i*2+1]]["derrotas"] += 1
                elif int(lista_gols[i][0]) < int(lista_gols[i][2]):
                    teams[lista_nome_times[i*2]]["derrotas"] += 1
                    teams[lista_nome_times[i*2+1]]["vitorias"] += 1
                else: 
                    teams[lista_nome_times[i*2]]["empates"] += 1
                    teams[lista_nome_times[i*2+1]]["empates"] += 1
            if lista_nome_times[i*2+1] == j:
                teams[j]["golsProFora"] += int(lista_gols[i][2])
                teams[j]["golsContraFora"] += int(lista_gols[i][0])
            
    print("\n")

for i in lista_nome_times:
    teams[i]["pontuacao"] = teams[i]["vitorias"]*3 + teams[i]["empates"]
    teams[i]["golsPro"] = teams[i]["golsProEmCasa"] + teams[i]["golsProFora"]
    teams[i]["golsContra"] = teams[i]["golsContraEmCasa"] + teams[i]["golsContraFora"]
    teams[i]["saldoDeGols"] = teams[i]["golsPro"] - teams[i]["golsContra"]
    teams[i]["saldoDeGolsEmCasa"] = teams[i]["golsProEmCasa"] - teams[i]["golsContraEmCasa"]
    teams[i]["saldoDeGolsFora"] = teams[i]["golsProFora"] - teams[i]["golsContraFora"]
```

```python
dataFortheDataFrame = []
for i in teams:
    dataFortheDataFrame.append(teams[i])

df = pd.DataFrame(dataFortheDataFrame)
df = df.sort_values(['pontuacao', 'saldoDeGols', 'id'], ascending=[False, False, True]).style.hide_index()
df
```
